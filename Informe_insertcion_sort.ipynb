#1. Descripción del problema

Un problema de ordenamiento de un conjunto de números

Entrada: una secuencia de números $[a_1,a_2,a_3,...,a_n]$

Salida: una permutación de la misma secuencia de números $[a_1,a_2,a_3,...,a_n]$ de tal forma que estos $a_1'\leq a_2'\leq a_3' \leq... \leq a_n'$.

#2. Insertcion sort

El insertcion sort es un algoritmo de ordenamiento el cual funciona haciendo comparaciones entre un elemento y sus antecesores

A continuación, pueden ver su implementación.


##2.1. Código

```python
import random
from termcolor import colored
import copy

def insertcion_sort(a,verbose):
  n = len(a) ##longitud del arreglo
  aux = 0 ##dato que auxilia al intercambio de datos
  c = 0 ##contador de comparaciones
  
  if(verbose):print("arreglo de entrada: ",a)

  for i in range(0,n): ##ciclo principal que se mueve a travez del arreglo
    k = i
    if(verbose):print("analisis de datos N°",i)
    while k > 0: ##ciclo que empieza a hacer las comparaciones
      if(verbose and k-1 != 0):
          print(str(a[:k-1])[1:-1],",",colored(f"{str(a[k-1:k+1])[1:-1]}","red"),",",str(a[k+1:])[1:-1],"\n")
      elif(verbose and k-1 == 0):
        print(colored(f"{str(a[k-1:k+1])[1:-1]}","red"),",",str(a[k+1:])[1:-1],"\n")
        
      if(a[k] < a[k-1]):
        aux = a[k-1]
        a[k-1] = a[k]
        a[k] = aux
        c = c + 1
        k = k - 1
        if(verbose and k != 0):
          print(str(a[:k])[1:-1],",",colored(f"{str(a[k:k+2])[1:-1]}","green"),",",str(a[k+2:])[1:-1],"\n")
        elif(verbose and k == 0):
          print(colored(f"{str(a[k:k+2])[1:-1]}","green"),",",str(a[k+2:])[1:-1],"\n")
      else: 
        c = c + 1
        break
    
    if(verbose):
      print("lo revisado esta ordenado")
      print(colored(f"{str(a[:i+1])[1:-1]}","blue"),",",str(a[i+1:])[1:-1],"\n")
    if(i == n and verbose):
      print("ya esta completamente ordenado")
    
  return a,c

#ejemplo
A = [30,12,2,6,21]
print("entrada: ", A)
A,cont = insertcion_sort(A,False)
print("salida: ", A)
print("numero de comparaciones: ", cont)
```

##2.2. Descripción del algoritmo

El algoritmo recibe de entrada una lista (o un arreglo) $a$ con una secuencia de $n$ elementos, estos elementos se ordenan dentro de la lista.

1. Al empezar asumimos que el primer elemento del arreglo `a[0]` ya está ordenado.

2. En cada iteración de $i$, el algoritmo se posiciona en el elemento `a[i]` y compara a este con su antecesor `a[i-1]`, si `a[i] < a[i-1]` estos elementos intercambian posición y $i$ decrece en una unidad.

3. Si la lista se encuentra ordenado, el algoritmo termina, En caso contrario se vuelve a repetir el paso 2.

4. Al finalizar el algoritmo retorna el arreglo ordenado y un contador del número de comparaciones realizadas para ordenarlo.

Cabe resaltar que a el algoritmo no le hace falta comparar todos los antecesores a un elemento para ordenarlo, este detalle lo hace más eficiente en comparación a otros algoritmos de ordenamiento como, por ejemplo: bubble sort.


##2.3. Ejemplo

Para ordenar el siguiente arreglo:

$[8,4,6,2]$

en la primera iteración del arreglo, se asume que el primero elemento esta ordenado.

$[8,4,6,2]$ -> $[8',4,6,2]$

en la segunda iteración se observa que el 4 es menor que el 8, por lo tanto, se intercambian.

$[8',4,6,2]$ -> $[4',8',6,2]$

en la tercera iteración se observa que el 6 es menor que el 8 pero no menor al 4, por lo tanto, este se intercambia únicamente con el 8.

$[4',8',6,2]$ -> $[4',6',8',2]$

en la cuarta iteración se observa que el 2 es menor que el 8, que el 6 y que el 4, por ello este intercambia se intercambiaría de posición con cada uno hasta quedar en la posición del 4.

$[4',6',8',2]$ -> $[4',6',2',8']$ -> $[4',2',6',8']$ -> $[2',4',6',8']$

# 4. Correctitud
## Teorema (Correctitud).
El algoritmo Insertion Sort retorna una lista con los mismos elementos de entrada, pero ordenados de menor a mayor.

## Prueba del teorema
Se puede observar que en cada iteración de Insertion Sort, se agrega un elemento ordenado antes del elemento en la posición *i*. Por lo que probaremos la siguiente propiedad invariante de bucle para el Insertion Sort:

> Al comienzo de cada iteración del primer bucle, los *i - 1* primeros elementos corresponden a elementos ordenados de menor a mayor.

## Inicialización

Al comienzo de la primera iteración (i = 1), los primeros *i - 1* elementos del arreglo cumplen esta propiedad. Se asume que el primer elemento ya está ordenado, por lo tanto, se cumple esta propiedad.

## Mantención

Aplicando inducción, asumimos que al comienzo de cada iteración se cumple la propiedad. Los *i - 1* primeros elementos del arreglo corresponden a elementos ordenados de menor a mayor.

Consideremos a X como el elemento que está en la posición *i* al comienzo de la *i*-ésima iteración.

Durante la iteración, X intercambia posiciones entre los elementos ya ordenados hasta que se cumpla la condición de que el valor del siguiente elemento sea menor que el valor de X.

Al finalizar la iteración, X queda en una posición menor a *i* y se seguirá cumpliendo la propiedad invariante.

## Correctitud
La propiedad de bucle invariante es verdadera al inicio del bucle y se mantiene en cada iteración. Por lo tanto, podemos decir que al finalizar la n-ésima iteración del algoritmo, el resultado será una lista con los mismos elementos del arreglo de entrada, pero ordenados de menor a mayor.

##5.1. Numero de comparaciones

En este experimento se comprobara el numero de comparaciones que el insertion sort necesita realizar para poder ordenar una lista en su totailidad.

Para ello se hizo una grafica en la que se muestra.

* La cantidad de comparaciones del peor caso calculadas matemáticamente,
* La cantidad de comparaciones del mejor caso calculadas matemáticamente (n-1 comparaciones si el arreglo está ordenado)
* La cantidad de comparaciones realizadas experimentalmente para entradas aleatorias

para un problema que tiene un tamaño $n \in [5,20]$

```python
import matplotlib.pyplot as plt

x=[n for n in range(5,21)] 
y1=[n*(n-1)/2 for n in range(5,21)] # worst case
y2=[n-1 for n in range(5,21)] # best case
y=[]; 

for n in range(5,21):
  a = random.sample(range(1, 100), n)
  a,counter = insertcion_sort(a,False)
  y.append(counter)

plt.plot(x,y)
plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["Insertcion Sort", "theoretical worst case", "theoretical best case"])

plt.xlabel('n')
plt.ylabel('number of operations')
plt.show()
```

###Análisis de resultados

Como se puede apreciar en el gráfico, el número de comparaciones necesarias cumple la expectativa, de realizar menos comparaciones que en el peor caso, pero no menos que en el mejor caso.

De hecho, el número de comparaciones necesarias está bastante cercano a lo que sería la mediana entre ambos casos.


##5.2. Tiempo de ejecución (experimental)

En el siguiente experimento se comprueba si el algoritmo cumple con el tiempo promedio esperado, que sería un tiempo $O(n^2)$.

```python
import matplotlib.pyplot as plt
import datetime
from timeit import repeat
import random

x=[]; y=[]

for n in range(5,500):

  a = random.sample(range(1, 1000), n)

  #la función repeat está ejecutando 20 veces insertcion_sort con el arreglo a y retorna el mejor tiepo obtenido.
  t = repeat(setup="from __main__ import insertcion_sort", stmt=f"insertcion_sort({a},{False})", repeat=1, number=10)

  x.append(n)
  y.append(t)


plt.plot(x,y)
plt.legend(["Insertcion Sort"])

plt.xlabel('n')
plt.ylabel('time in ms')
plt.show()
```
